```python
def gradeevaluation(grade):
    if grade < 0 or grade > 100:
        return "Invalid grade"
    elif grade >= 0 and grade < 60:
        return "F"
    elif grade >= 60 and grade < 70:
        return "D"
    elif grade >= 70 and grade < 80:
        return "C"
    elif grade >= 80 and grade < 90:
        return "B"
    elif grade >= 90 and grade <= 100:
        return "A"
```

```python
def coins(value):
    if value <= 0 or value>= 100:
        return "Invalid number"
    value = int(value)
    values = [value, int(value % 25), (int(value % 25)) % 10, int(int(value % 25) % 10) % 5]
    coins = [int(values[0]/25), int(values[1]/10), int(values[2]/5), int(values[3])]
    coinnames = [" quarter", " dime", " nickel", " penny"]
    coinsnames = [" quarters", " dimes", " nickels", " pennies"]
    text = list()
    finaltext=""
    for i in range(len(coins)) : 
        if coins[i] == 1:
            text.append(str(coins[i])+ coinnames[i])
        elif coins[i] > 1:
            text.append(str(coins[i])+ coinsnames[i])
    finaltext=text[0]
    if len(text) > 2:
        for i in range(1,len(text)-1):
            finaltext = finaltext + ", " + text[i]
        finaltext = finaltext + " and " + text[len(text)-1] + "."
    if len(text) == 2:
        finaltext = finaltext + " and " + text[len(text)-1] + "."
    if len(text) == 1:
        finaltext = finaltext + "."
    return finaltext
        
```

```python
for i in range(1,100):
    print(coins(i))
```

```python
def GCD(num1,num2):
    num1 = int(num1)
    num2 = int(num2)
    list1 = [i for i in range(1,num1+1) if num1 % i == 0]
    list2 = [i for i in range(1,num2+1) if num2 % i == 0]
    divisor = 1
    for i in list1:
        for j in list2:
            if i == j:
                divisor = j
    return divisor
```

```python
print(GCD(93,76))
```

```python
def LCM(num1,num2):
    num1 = int(num1)
    num2 = int(num2)
    i = 1
    if num1 >= num2:
        while num1 * i <= num1 * num2:
            if num1 * i % num2 == 0:
                return num1 * i
            i = i + 1
    else:         
        while num2 * i <= num1 * num2:
            if num2 * i % num1 == 0:
                return num2 * i
            i = i + 1
```
